#sentemnt analysis



```python
Reviews= df[['product_id','review_title','review_content']]
```

```python
!pip install twython
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
nltk.download('vader_lexicon')
model = SentimentIntensityAnalyzer()
```

```python
result = {}

for i, row in tqdm(Reviews.iterrows(), total=len(Reviews)):
    review_desc =row['review_content']
    myid = row['product_id']
    result[myid] = model.polarity_scores(review_desc)
```

```python
pol_score = pd.DataFrame(result).T
pol_score = pol_score.reset_index().rename(columns={'index' : 'product_id'})

pol_score
```

```python
final_df=df.merge(pol_score,how='left')
final_df
```
